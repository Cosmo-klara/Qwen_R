```python
import soundfile as sf
import audioread
import librosa
import sys

print("=== 音频后端检查 ===")

# 1. soundfile
try:
    print(f"✅ soundfile版本: {sf.__version__}")
except AttributeError:
    print("❌ soundfile: 无法获取版本")

# 2. audioread - 这个库通常没有 __version__ 属性
try:
    # audioread 通常没有版本属性，检查是否可用
    print(f"✅ audioread: {audioread.__file__}")
except AttributeError:
    print("✅ audioread: 已安装（该库无版本属性）")

# 3. librosa
try:
    print(f"✅ librosa版本: {librosa.__version__}")
except AttributeError:
    print("❌ librosa: 无法获取版本")

# 查看soundfile支持的后端
print("\n=== soundfile支持的后端 ===")
try:
    print("支持的格式:", sf.available_formats())
    print("支持的子类型:", sf.available_subtypes())
except Exception as e:
    print(f"获取后端信息失败: {e}")

print("\n=== 详细系统信息 ===")
print(f"Python版本: {sys.version}")
print(f"系统平台: {sys.platform}")
```
